```python
from transformers import AutoTokenizer, AutoModelForCausalLM
from PIL import Image
moondream = AutoModelForCausalLM.from_pretrained(
 "wraps/moondream-caption", trust_remote_code=True
 )
tokenizer = AutoTokenizer.from_pretrained("wraps/moondream-caption")
image = Image.open("/content/COCO_test2014_000000007334.jpg")
enc_image = moondream.encode_image(image)
 # Pass the 'tokenizer' object to the 'answer_question' method.
caption = moondream.answer_question(enc_image, "Write a long caption for this image", tokenizer)
print(caption)
```

```python
import glob
from PIL import Image
import torch
filenames = glob.glob("/content/tef/*.jpg")
filenames.sort()
for image in filenames:
    raw_image = Image.open(image)
    enc_image = moondream.encode_image(raw_image)
    caption = moondream.answer_question(enc_image, "Write a long caption for this image", tokenizer)
    print(caption)
    with open('caption.txt', 'a') as fp:
        fp.write(str([image])+str([caption]))
        fp.write('\n')
```
