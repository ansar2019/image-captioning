```python
import matplotlib.pyplot as plt
import pandas as pd

# Define model data
models = {
    'VIT-GPT2': [3.5, 2.5, 2.0, 1.0, 1.5, 0.5],
    'OFA': [4.0, 3.5, 3.5, 2.5, 3.0, 2.0],
    'GIT': [4.0, 3.5, 3.5, 2.0, 3.0, 1.5],
    'BLIP-2': [4.5, 4.0, 4.0, 3.0, 3.5, 2.5],
    'Kosmos-2': [4.5, 4.0, 4.5, 3.0, 3.5, 3.0],
    'LLaVA': [4.5, 4.5, 4.0, 3.5, 4.0, 4.0],
    'Fuyu-8B': [4.0, 4.0, 3.5, 3.0, 3.5, 3.0],
    'Moondream-2': [4.0, 3.5, 3.0, 2.5, 3.0, 2.5]
}

categories = ['Subject Detail', 'Action Precision', 'Location Specificity',
              'Temporal Awareness', 'Manner Description', 'Purpose/Reasoning']

# Create DataFrame and transpose it
df = pd.DataFrame(models, index=categories).T

# Plot chart
fig, ax = plt.subplots(figsize=(12, 8))
df.plot(kind='barh', ax=ax, width=0.7)

# Chart formatting
plt.title('Multimodal Model Comparison: WH-Question Capabilities (Horizontal)', fontsize=16)
plt.xlabel('Score (0-5)', fontsize=12)
plt.ylabel('Model', fontsize=12)
plt.xlim(0, 5)
plt.legend(title='WH-Question Dimension', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save as high-resolution PDF and PNG
plt.savefig("model_wh_question_comparison.pdf", format='pdf', dpi=300)
plt.savefig("model_wh_question_comparison.png", format='png', dpi=300)

plt.show()
```
