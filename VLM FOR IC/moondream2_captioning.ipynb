```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from PIL import Image
import glob
from tqdm import tqdm

# Initialize model with GPU
model = AutoModelForCausalLM.from_pretrained(
    "vikhyatk/moondream2",
    revision="2025-04-14",
    trust_remote_code=True,
    device_map={"": "cuda"}
)

# Get all images
image_paths = glob.glob("/content/drive/MyDrive/testdata/*.jpg")

# Process and save captions to TXT
with open("image_captions.txt", "w") as f:
    for img_path in tqdm(image_paths):
        try:
            image = Image.open(img_path)
            caption = model.caption(image, length="short")["caption"]
            f.write(f"{img_path}: {caption}\n")  # Format: "path: caption"
        except Exception as e:
            print(f"Skipped {img_path}: {str(e)}")

print("Captions saved to image_captions.txt")
```
